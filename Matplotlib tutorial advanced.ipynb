###### Simple plot

```python
import pylab as pl
import numpy as np

#World historical and predicted populations by percentage distribution
africa = [18.8, 19.7, 15.5, 13.4, 10.9, 8.8, 8.1, 8.8, 12.8, 14.5, 14.8, 15.2, 19.8, 23.7]
europe = [18.3, 19.1, 18.3, 20.6, 20.8, 21.9, 24.7, 21.7, 12.2, 10.9, 10.7, 10.5, 7.0, 5.3]
year = [1500, 1600, 1700, 1750, 1800, 1850, 1900, 1950, 1999, 2008, 2010, 2012, 2050, 2150]
pl.plot(africa)
pl.plot(europe)
pl.show()
```

###### Pretty plot

```python
#Create a figure of size 8x6 inches, 80 dots per inch
pl.figure(figsize=(8, 6), dpi=80)

#Set x and y limits for clearer display
pl.xlim(1500, 2150)
pl.ylim(4, 25)

#Set x and y ticks
pl.xticks(np.linspace(1500, 2150, 6, endpoint=True))
pl.yticks(np.linspace(5.0, 25.0, 11, endpoint=True))

#Add labels
plt.xlabel("Year", size=20)
plt.ylabel("Percentage of world\ntotal population", size=20)
plt.title("World historical and predicted populations\nby percentage distribution", size=20)

#Adjust appearance of lines, give them labels and plot
pl.plot(year, africa, color="red", linewidth=2.0, linestyle="-", label = "africa")
pl.plot(year, europe, color="green", linewidth=2.0, linestyle="-", label = "europe")

pl.plot([1617, 1617], [0, 18.8], color='orange', linewidth=1.5, linestyle="--")
pl.annotate("Europe's percentage of world\npopulation surpasses Africa's",
            xy=(1617, 18.8), xycoords='data',
            xytext=(-50, +50), textcoords='offset points', fontsize=12,
            arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))


pl.plot([1997, 1997], [0, 12.5], color='orange', linewidth=1.5, linestyle="--")
pl.annotate("Africa's percentage of world\npopulation surpasses Europe's",
            xy=(1997, 12.5), xycoords='data',
            xytext=(+25, +50), textcoords='offset points', fontsize=12,
            arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))

#Add legend
pl.legend(loc='lower left')

#Adjust spines. Top and right are removed.
ax = pl.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

#Make labels more visible
for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_fontsize(16)
    label.set_bbox(dict(facecolor='white', edgecolor='None', alpha=0.65))
    
#Display plot
pl.show()
```
